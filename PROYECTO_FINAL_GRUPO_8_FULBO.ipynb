```python
#Goles que hizo cada pais cuando estaba de local
data_local = data.groupby("team1")["number of goals team1"].sum()
data_local = pd.DataFrame(data=data.groupby("team1")["number of goals team1"].sum())
data_local
```

```python
#Goles que hizo cada pais cuando estaba de visitante
data_visitante = data.groupby("team2")["number of goals team2"].sum()
data_visitante = pd.DataFrame(data=data.groupby("team2")["number of goals team2"].sum())
data_visitante
```

```python
#Goles que hizo cada pais en todo el torneo
goles_local = data.groupby("team1")["number of goals team1"].sum()
goles_visitante = data.groupby("team2")["number of goals team2"].sum()
goles = goles_local+goles_visitante
goles.index.name = "Paises"
sns.barplot (x=goles, y=goles.index)
```

```python
goles0 = pd.DataFrame(goles)
goles0.columns = ['Goles']
goles0
```

```python
#Reemplazamos los valores de posesion de str a float
posesiones1 = data['possession team1'].str.replace("%","").astype(float)
posesiones2 = data['possession team2'].str.replace("%","").astype(float)
posesion_disputada = data['possession in contest'].str.replace("%","").astype(float)
posesiones1
```

```python
#Promedio de Posesion de los paises cuando eran locales
posesion_local = data.groupby("team1")["possession team1"].mean()
sns.barplot(x=posesion_local, y=posesion_local.index)
```

```python
#Una Tabla para tener el promedio de cada pais su valor en concreto
posesion_local0 = pd.DataFrame(posesion_local)
posesion_local0
```

```python
#Ahora el pomedio de posesion de la pelota cuando eran visitantes
posesion_visitante = data.groupby("team2")["possession team2"].mean()
sns.barplot(x=posesion_visitante, y=posesion_visitante.index)
```

```python
posesion_visitante0 = pd.DataFrame(posesion_visitante)
posesion_visitante0
```

```python
#Promedio de posesion que tuvieron todos los paises en todo el torneo
Posesion_Total = (posesion_visitante + posesion_local)/2
Posesion_Total.index.name = "Paises"
sns.barplot(x=Posesion_Total, y=Posesion_Total.index)
```

```python
Posesion_Total0 = pd.DataFrame(Posesion_Total)
Posesion_Total0.columns = ['Posesiones']
Posesion_Total0
```

```python
#Oportunidades que tuvo cada Pais
intentos1 = data.groupby("team1")["total attempts team1"].sum()
intentos2 = data.groupby("team2")["total attempts team2"].sum()
Intentos = intentos1 + intentos2
Intentos.index.name = "Paises"
sns.barplot(x=Intentos, y=Intentos.index)
```

```python
Intentos0 = pd.DataFrame(Intentos)
Intentos0.columns = ['Oportunidades']
Intentos0 #Argentina le puso mas huevos
```

```python
#Tarjetas amarillas cobradas a cada pais
amarillas1 = data.groupby("team1")["yellow cards team1"].sum()
amarillas2 = data.groupby("team2")["yellow cards team2"].sum()
amarillas = amarillas1 + amarillas2
amarillas.index.name = "Paises"
sns.barplot(x=amarillas, y=amarillas.index)
```

```python
amarillas0 = pd.DataFrame(amarillas)
amarillas0.columns = ['Tarjetas']
amarillas0
```

```python
#Tarjetas rojas cobradas a cada pais. Los equipos mas brutos💀
rojas1 = data.groupby("team1")["red cards team1"].sum()
rojas2 = data.groupby("team2")["red cards team2"].sum()
rojas = rojas1 + rojas2
rojas.index.name = "Paises"
sns.barplot(x=rojas, y=rojas.index)
```
